## Predict Using Linear Regression

[Simple Linear Regression](https://en.wikipedia.org/wiki/Simple_linear_regression#Fitting_the_regression_line)

$$\hat{y} = \beta_0 + \beta_1 x$$

where

> $$ \beta_1 = \frac{\sum_{i=1}^n (y_i - \bar{y} ) (x_i - \bar{x} )}{\sum_{i=1}^n (x_i - \bar{x})^2} $$

and

> $$ \beta_0 = \bar{y} - \beta_1\bar{x} $$

Given variables:
- $\bar{y}$ : the sample mean of observed values $Y$
- $\bar{x}$ : the sample mean of observed values $X$
- $s_Y$ : the sample standard deviation of observed values $Y$
- $s_X$ : the sample standard deviation of observed values $X$
- $r_{XY}$ : the sample Pearson correlation coefficient between observed $X$ and $Y$

### Define X, Y

- X = Predictor Variable 
- Y = Target Variable

```python
X = house['sq__ft']
Y = house['price']
```

### Calculate beta_1

$$ \beta_1 = \frac{\sum_{i=1}^n (y_i - \bar{y} ) (x_i - \bar{x} )}{\sum_{i=1}^n (x_i - \bar{x})^2} $$

```python
def calculate_beta_1(X,Y):
    
    Xbar = np.mean(X)
    Ybar = np.mean(Y)
    
    n = np.sum((Y-Ybar)*(X-Xbar))
    d = np.sum(np.square(X-Xbar))
    
    beta_1 = n/d
    
    return beta_1
```

```python
beta_1 = calculate_beta_1(X,Y)
beta_1
```

### Calculate beta_0

$$ \beta_0 = \bar{y} - \beta_1\bar{x} $$

```python
def calculate_beta_0(X,Y,beta_1):
    Xbar = np.mean(X)
    Ybar = np.mean(Y)
    
    beta_0 = Ybar - beta_1*Xbar
    
    return beta_0
```

```python
beta_0 = calculate_beta_0(X,Y,beta_1)
beta_0
```

### Find Yhat (Predicted Value Using Beta 0 and Beta 1)

$$\hat{y} = \beta_0 + \beta_1 x$$

```python
def predict_Yhat(X,beta_1,beta_0):
    Yhat = beta_0 + beta_1*X
    
    return Yhat
```

```python
Yhat = predict_Yhat(X,beta_1,beta_0)
Yhat
```

### Find Residuals

$$ \text{residual}_i = y_i - \hat{y}_i$$

```python
def find_residuals(Yhat,Y):
    residual = Yhat - Y
    
    return residual
```

```python
residual = find_residuals(Yhat,Y)
residual
```

### Find R-Square

The coefficient $R^2$ is defined as $$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$, where $$ SS_{res} $$ is the residual sum of squares $$ SS_{res} = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$ and $$ SS_{res} $$ is the total sum of squares $$ SS_{tot} = \sum_{i=1}^n \left(y_i - \bar{y}\right)^2 $$

The best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse).

**Write a function to calculate R- Square value.**

The total sum of squares:

$$ SS_{tot} = \sum_{i=1}^n \left(y_i - \bar{y}\right)^2 $$

The residual sum of squares:

$$ SS_{res} = \sum_{i=1}^n \left(y_i - \hat{y}_i\right)^2 $$

$R^2$:

$$ R^2 = 1 - \frac{SS_{res}}{SS_{tot}} $$

```python
def find_rsquare(Yhat,Y):
    Ybar = np.mean(Y)
    sumsqtot = np.sum((Y-Ybar)**2)
    sumsqres = np.sum((Y-Yhat)**2)
    rsquare = 1 - sumsqres/sumsqtot
    
    return rsquare
```

```python
rsquare = find_rsquare(Yhat,Y)
rsquare
```
