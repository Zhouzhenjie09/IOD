```python
googlenews=GoogleNews(start='11/9/2021',end='11/03/2022')
googlenews.search('Bitcoin')#, 'Cryptocurrency', 'Crypto')
result=googlenews.result()
for i in range(2,30):
    googlenews.getpage(i)
    result=googlenews.result()
    df=pd.DataFrame(result)
#df=pd.DataFrame(result)
#df.head()
```
