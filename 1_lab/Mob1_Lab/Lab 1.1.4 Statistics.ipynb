# [BONUS] Plot X, Y and Best Fit Line

**Write a function to calculate best fit.**

![alt text](https://i.imgur.com/B43vOV0.jpg)[1]

[1]: https://www.mathsisfun.com/data/least-squares-regression.html "Least Squares Regression"

````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    '''
      
    # To Do
    x_sq = [x**2 for x in X]
    xy = [ x*y for x,y in zip(X,Y)]
    x_sum = sum(X)
    y_sum = sum(Y)
    x_sq_sum = sum(x_sq)
    xy_sum = sum(xy)
    
    n = len(X)
    
    m = ((n*xy_sum)-(x_sum*y_sum)) / ((n*x_sq_sum) - (x_sum**2))
    
    b = (y_sum - (m* x_sum)) / n
    print(m,b)
    
    print('best fit line:\ny = {:.2f}x + {:.2f}'.format(m, b))
    return m, b

# solution
m, b = best_fit(X, Y)
#best fit line:
#y = mx + b
#print(y)


# # To Do: plot points and fit line
mx = [m*x for x in X]
print(mx)

y_fit = []
for z in range(0, len(mx)):
    y_fit.append(float(mx[z]) + float(b[z]))
#y_fit = [mx + [b]]
print(y_fit)

#y_fit = [m + b * x for x in X]

plt.scatter(X,Y)
plt.plot(X,y_fit )

````
